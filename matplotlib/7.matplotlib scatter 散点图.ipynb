```python
import matplotlib.pyplot as plt
```

```python
import numpy as np
```

```python
plt.scatter(np.arange(5),np.arange(5))
plt.show()
```

```python
x = np.random.normal(0,1,500)
y = np.random.normal(0,1,500)

# c颜色 alpha 透明度 0-1 之间
plt.scatter(x,y,s=50,c='b',alpha=1)
plt.xlim((-2,2))
plt.ylim((-2,2))

plt.xticks(())
plt.yticks(())
plt.show()
```
